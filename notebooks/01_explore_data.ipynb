# 01: Explore the DSC-1 Spectral Unity Dataset

This notebook provides a guided tour of the complete experimental dataset.

**Contents:**
1. Load the data
2. Overview of all experiments
3. Riemann zeros and Monster-Zeta mapping
4. Four Walls of Impossibility
5. Verification scorecard

```python
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load the complete dataset
with open('../data/spectral-unity/experiment_results.json') as f:
    data = json.load(f)

print('Dataset keys:', list(data.keys()))
print(f'Total experiments: {len(data)}')
```

## Dataset Overview

The dataset contains results from 10 GPU-verified experiments spanning number theory, random matrix theory, algebraic geometry, and mathematical logic.

```python
# Quick summary of each section
for key in data:
    val = data[key]
    if isinstance(val, dict):
        print(f"\n{key}: {len(val)} fields")
        for k, v in val.items():
            if isinstance(v, list):
                print(f"  {k}: list of {len(v)} items")
            elif isinstance(v, dict):
                print(f"  {k}: dict with {len(v)} entries")
            else:
                print(f"  {k}: {v}")
    elif isinstance(val, list):
        print(f"\n{key}: list of {len(val)} items")
```

## Riemann Zeros

The first 50 non-trivial zeros of the Riemann zeta function, with the Monster-Zeta mapping $\Phi(n) = 2\pi\gamma_n / \log|M|$.

```python
zeros = data['riemann_zeros']['zeros']
phi_mapping = data['riemann_zeros']['phi_mapping']
lambda_M = data['riemann_zeros']['monster_wavelength']

print(f'Number of zeros: {len(zeros)}')
print(f'First zero: gamma_1 = {zeros[0]}')
print(f'Last zero: gamma_50 = {zeros[-1]}')
print(f'Monster wavelength: lambda_M = {lambda_M}')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.scatter(range(1, 51), zeros, c='blue', s=20)
ax1.set_xlabel('Zero index n')
ax1.set_ylabel('Imaginary part gamma_n')
ax1.set_title('First 50 Riemann Zeta Zeros')
ax1.grid(True, alpha=0.3)

ax2.scatter(range(1, 51), phi_mapping, c='purple', s=20)
ax2.set_xlabel('Zero index n')
ax2.set_ylabel('Phi(n) = 2*pi*gamma_n / log|M|')
ax2.set_title('Monster-Zeta Mapping')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Four Walls of Impossibility

Each wall has a characteristic frequency, harmonic number, and horizon entropy.

```python
walls_df = pd.DataFrame({
    'Wall': data['four_walls']['walls'],
    'Frequency (Hz)': data['four_walls']['frequencies'],
    'Harmonic': data['four_walls']['harmonics'],
    'Entropy': data['four_walls']['entropies'],
    'Ratio to Monster': data['four_walls']['ratios_to_monster'],
})

print(walls_df.to_string(index=False))
```

## Verification Scorecard

Summary of all 7 framework predictions.

```python
verification_df = pd.DataFrame([
    {'Prediction': k, 'Status': v['status'], 'Result': v['result']}
    for k, v in data['verification'].items()
])

print(verification_df.to_string(index=False))

# Count by status
status_counts = verification_df['Status'].value_counts()
print(f'\nSupported: {len(verification_df[verification_df["Status"] != "NOT SUPPORTED"])}/7')
print(f'Not supported: {len(verification_df[verification_df["Status"] == "NOT SUPPORTED"])}/7')
```
